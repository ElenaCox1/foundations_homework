```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.epicurious.com/search/rhubarb')
doc = BeautifulSoup(response.text)
```

```python
rows =[]
recipes = doc.find_all("article")
for recipe in recipes:
    row = {}
    if recipe['class'] == ['legal-notice']:
        pass
    else:
        row['tag'] = recipe.find(class_="tag").text
    try: 
        row['recipe name'] = recipe.find(class_="hed").text
    except:
        pass   
    try: 
        row['summary'] = recipe.find(class_="dek").text
    except:
        pass
    try: 
        row['rating'] = recipe.find(itemprop="ratingValue").text
    except:
        pass
    try:     
        row['would make again'] = recipe.find(class_="make-again-percentage").text
    except:
        pass
    try: 
        row['URL'] = recipe.find('header').find('a')['href']
    except:
        pass
    rows.append(row)
rows
```

```python
def scrape_page(row):
    url= f"https://www.epicurious.com{row['URL']}"
    print("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)

    page = {}
    try:
        page['ingredients'] = doc.find(class_='ingredients-info').text
    except:
        pass
    try:
        page['directions'] = doc.find(class_='instructions').text
    except:
        pass
    try:
        page['tags'] = doc.find(class_='tags').text
    except:
        pass
    return pd.Series(page)
```

```python
scraped_df = df.head().apply(scrape_page, axis=1)
```
